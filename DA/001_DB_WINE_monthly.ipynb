###**2.1.1 | CORRECT INF VALUES - COUNTRY CHECK**

```python
df.loc[(df['EUR_KG'] == np.inf) & (df['EUR'] != 0), 'EUR_KG'] = np.nan
```

```python
df[df['EUR_KG'].isna() & (df['EUR'] != 0)]
```

The missing values on TON columns will be solved by:
1. Doing the average price (EUR_TON) from the previous and next month. Then apply the calculus to the TON column later.
2. If is missing any other value, just will be take the previous or next value.

```python
mean_values = df.groupby(['country', 'year'])['EUR_KG'].transform('mean')
```

```python
df.loc[df['EUR_KG'].isna() & (df['EUR'] != 0), 'EUR_KG'] = mean_values
```

###**2.1.3 | CORRECT MISSING VALUES IN df['TON'] & ROUND**

```python
df['TON'] = df['EUR'] / df['EUR_KG']
```

```python
df['EUR_KG'] = round(df['EUR_KG'],3)
```

###**2.1.4 | LAST AMENDS**

Finally there are still 4 registers that shows null values. This is because there were no other monthly data on the NaN value year. In this case, it will be done a general period mean to input the values.

```python
df[df['TON'].isna() & (df['EUR'] != 0)]
```

```python
mean_values2 = df.groupby(['country'])['EUR_KG'].transform('mean')
```

```python
df.loc[df['EUR_KG'].isna() & (df['EUR'] != 0), 'EUR_KG'] = mean_values2
```

```python
df['TON'] = df['EUR'] / df['EUR_KG']
df['EUR_KG'] = round(df['EUR_KG'],3)
```
